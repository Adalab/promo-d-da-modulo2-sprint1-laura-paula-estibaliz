Moda, Media, Mediana, Máximo, Mínimo de la columna 'averageprice' y 'total_bags'

```python
media = df_avocado['averageprice'].mean()
print(f' La media de la columna "averageprice" es {media}')
print('---------')

mediana = df_avocado['averageprice'].median()
print(f' La mediana de la columna "averageprice" es {mediana}')
print('---------')

moda = df_avocado['averageprice'].mode()[0]
print(f' La moda de la columna "averageprice" es {moda}')
print('---------')

max_price = df_avocado['averageprice'].max()
print(f' El valor máximo de la columna "averageprice" es {max_price}')
print('---------')

min_price = df_avocado['averageprice'].min()
print(f' El valor mínimo de la columna "averageprice" es {min_price}')
```

```python
media = df_avocado['total_bags'].mean()
print(f' La media de la columna "total_bags" es {media}')
print('---------')

mediana = df_avocado['total_bags'].median()
print(f' La mediana de la columna "total_bags" es {mediana}')
print('---------')

moda = df_avocado['total_bags'].mode()[0]
print(f' La moda de la columna "total_bags" es {moda}')
print('---------')

max_price = df_avocado['total_bags'].max()
print(f' El valor máximo de la columna "total_bags" es {max_price}')
print('---------')

min_price = df_avocado['total_bags'].min()
print(f' El valor mínimo de la columna "total_bags" es {min_price}')
```

Sacamos la tabla de contingencia para las dos variables categóricas que vamos a generar a partir del precio medio y las bolsas totales.

```python
df_avocado['averageprice_bins'] = pd.cut(df_avocado['averageprice'], bins = (0, 1.50, 2.5, 3.26), labels= ['Bajo', 'Medio', 'Alto'])
df_avocado['total_bags_bins'] = pd.cut(df_avocado['total_bags'], bins = (0, 60000, 110000, 194000), labels= ['Bajo', 'Medio', 'Alto'])
```

```python
df_avocado_crosstab = pd.crosstab(df_avocado['averageprice_bins'],
                                df_avocado['total_bags_bins'])
df_avocado_crosstab
```
