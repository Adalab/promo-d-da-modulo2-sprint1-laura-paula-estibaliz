2. Filtrad los datos para quedarnos solo con los datos de la categor√≠a "Software".

```python
# extraimos solo los datos de la categor√≠a software

categor√≠a_soft = df_2016[df_2016["categor√≠a"] == 'Software']
categor√≠a_soft
```

```python
# extraimos solo los datos de la categor√≠a hardware

categor√≠a_hard = df_2016[df_2016["categor√≠a"] == 'Hardware']
categor√≠a_hard
```

   
   2.1 Dentro del conjunto de datos filtrado que acabamos de generar, realiza la mejor visualizaci√≥n para mostrar cuantos tickets tenemos de cada tipo que tenemos (Problema o Solicitud).

```python
# extraimos los datos de cada tipo de ticket (problema y solicitud)

df_count_soft = pd.DataFrame(categor√≠a_soft["tipo"].value_counts())
df_count_soft.reset_index(inplace=True)


df_count_soft
```

```python
# creamos el √≠ndice del df con el m√©todo reset.index()

df_count_hard = pd.DataFrame(categor√≠a_hard["tipo"].value_counts())
df_count_hard.reset_index(inplace=True)
df_count_hard
```

```python
# creamos la gr√°fica de tipos para cada ticket (problema y solicitud) de la catefor√≠a software 

plt.figure()
plt.bar(df_count_soft["index"], df_count_soft["tipo"])
plt.show()
```

```python
# creamos la gr√°fica de tipos para cada ticket (problema y solicitud) de la catefor√≠a hardware

plt.figure()
plt.bar(df_count_hard["index"], df_count_hard["tipo"])
plt.show()
```

Las solicitudes para software son mas que el doble de las solicitudes para hardware. La variable problema se porta de la misma forma. 

  2.2 Usando el mismo conjunto de datos filtrado, ¬øexiste alguna relaci√≥n entre la satisfacci√≥n y los d√≠as de resoluci√≥n?. Poned los nombres de los ejes, siendo el eje x los "d√≠as de resoluci√≥n" y el eje y la "satisfacci√≥n".

```python
# creamos la gr√°fica de satisfacci√≥n por d√≠as_resoluc√≠on para la categor√≠a software

plt.figure(figsize = (10,5))

plt.hist(categor√≠a_soft["satisfacci√≥n"], color = "green")

plt.xlabel("d√≠as_resoluci√≥n")

plt.ylabel("satisfacci√≥n")

plt.title("satisfacci√≥n_d√≠as_resoluci√≥n")

plt.grid(color = "grey", linestyle = "-.")

plt.show()
```

```python
# creamos la gr√°fica de satisfacci√≥n por d√≠as_resoluc√≠on para la categor√≠a hardware

plt.figure(figsize = (10,5))

plt.hist(categor√≠a_hard["satisfacci√≥n"], color = "green")

plt.xlabel("d√≠as_resoluci√≥n")

plt.ylabel("satisfacci√≥n")

plt.title("satisfacci√≥n_d√≠as_resoluci√≥n")

plt.grid(color = "grey", linestyle = "-.")

plt.show()
```

- Se puede notar que en general, para las dos categor√≠as, cuanto m√°s dias de resolucion (hasta 5 d√≠as) mas satisfechos est√°n los clientes. 
- En ambos los casos, los clientes cuyos tickets tardan 2 d√≠as para la resolucion no suelen estar satisfechos.
