2. Desde la direcci칩n quieren seguir evaluando la situaci칩n planteada en el ejercicio anterior. Pero ahora quieres saber cual es la media de la satisfacci칩n por categor칤a. En concreto quieren saber si independientemente de la categor칤a los clientes salgan satisfechos, para que en caso de que no sea as칤 poder identificar un posible problema subyacente. Por lo tanto, nos piden que calculemos la media de satisfacci칩n por cada una de las categor칤as que tenemos en el dataframe.

```python
df_satisfaccion = df_historico.groupby(['categor칤a'])['satisfacci칩n'].mean().reset_index()
df_satisfaccion
```

3. Quieren mejorar la consulta anterior y ahora nos piden que adem치s de la media les devolvamos la desviaci칩n est치ndar de la satisfacci칩n.
En nuestros datos, adem치s hemos visto que no solo tenemos categor칤as, y que las diferencias entre ellas no son especialmente significativas.
```python
df_satisfaccion = df_historico.groupby(['categor칤a'])['satisfacci칩n'].agg(['mean', 'std']).reset_index()
df_satisfaccion
```

游눠Pista Para responder a esta pregunta tendremos que unir los dataframes con m칠todos aprendidos en lecciones anteriores de Pandas.


7. BONUS Por 칰ltimo, quieres conocer los 5 empleados que tienen mayor valor en la satisfacci칩n y los 5 empleados que tienen un menor valor de satisfacci칩n. Todo esto respecto a la media. En este caso, conseguiremos una dataframe, donde tengamos el el id del empleado y su satisfacci칩n, pero no sabemos su nombre y appelidos. 쯉e os ocurre como conseguir la informaci칩n del nombre del empleado?

```python
df_nivel_max_sat = df_media_bonus.groupby(['departamento', 'id_empleado'])['satisfacci칩n'].mean().sort_values(ascending=False).reset_index()
df_nivel_max_sat.head(5)
```

```python
df_nivel_min_sat = df_media_bonus.groupby(['departamento', 'id_empleado'])['satisfacci칩n'].mean().sort_values(ascending=False).reset_index()
df_nivel_min_sat.tail(5)
```

```python
df_nombres_max = df_media_bonus.groupby(['departamento', 'id_empleado', 'nombre', 'apellido'])['satisfacci칩n'].mean().sort_values(ascending=False).reset_index()
df_nombres_max.head(5)
```

```python
df_nombres_min = df_media_bonus.groupby(['departamento', 'id_empleado', 'nombre', 'apellido'])['satisfacci칩n'].mean().sort_values(ascending=False).reset_index()
df_nombres_min.tail(5)
```
