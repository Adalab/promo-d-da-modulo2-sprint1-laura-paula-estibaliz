```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_puestos = pd.read_csv("datos/puestos_empleados.csv")
df_puestos.head()
```

```python
df_historico = pd.read_csv("datos/historico_tickets.csv")
df_historico.head()
```

```python
df_puestos.drop(['Unnamed: 0'], axis = 1, inplace = True)
df_puestos
```

```python
df_historico.drop(['Unnamed: 0'], axis = 1, inplace = True)
df_historico
```

1. En pairs anteriores hemos ido pudiendo sacar algunas conclusiones y contestando las preguntas que nos hacían las jefas. Desde la dirección se están planteando cambiar el sistema de asiganción de las solicitudes que llegan. En concreto, necesitan saber que categoría es la que recibe más tickets para poder asignar más empleados a esa categoría. Por esto nos piden que averiguemos qué categoría es la que tiene más tickets. Además nos piden que devolvamos esta consulta en un DataFrame.

📌 NOTA Ordena los resultados de mayor a menor. Esto todavía no lo hemos visto, pero un DataFrame lo podemos ordenar con .sort_values(). Aquí tenéis una explicación de cómo hacerlo.

```python
df_tickets = df_historico.groupby('categoría')['id_ticket'].count().sort_values(ascending= False).reset_index()
df_tickets
```

2. Desde la dirección quieren seguir evaluando la situación planteada en el ejercicio anterior. Pero ahora quieres saber cual es la media de la satisfacción por categoría. En concreto quieren saber si independientemente de la categoría los clientes salgan satisfechos, para que en caso de que no sea así poder identificar un posible problema subyacente. Por lo tanto, nos piden que calculemos la media de satisfacción por cada una de las categorías que tenemos en el dataframe.

```python
df_satisfaccion = df_historico.groupby(['categoría'])['satisfacción'].mean().reset_index()
df_satisfaccion
```

3. Quieren mejorar la consulta anterior y ahora nos piden que además de la media les devolvamos la desviación estándar de la satisfacción.
En nuestros datos, además hemos visto que no solo tenemos categorías, y que las diferencias entre ellas no son especialmente significativas. Revisando los datos de nuevo, desde dirección saben que hay dos tipos de tickets (Problema y Solicitud), basándonos en esto, ¿hay diferencias en la satisfacción basándonos en la categoría y el tipo de tickets?.

```python
df_satisfaccion = df_historico.groupby(['categoría'])['satisfacción'].agg(['mean', 'std']).reset_index()
df_satisfaccion
```

4. En nuestros datos, además hemos visto que no solo tenemos categorías, y que las diferencias entre ellas no son especialmente significativas. Revisando los datos de nuevo, desde dirección saben que hay dos tipos de tickets (Problema y Solicitud), basándonos en esto, ¿hay diferencias en la satisfacción basándonos en la categoría y el tipo de tickets?.

```python
df_satisfaccion = df_historico.groupby(['categoría', 'tipo'])['satisfacción'].agg(['mean', 'std']).reset_index()
df_satisfaccion
```


5. A pesar de que sabemos que tenemos muchos empleados y varios departamentos, nuestras jefas no saben cuántas personas hay en cada uno de estos departamentos. Por esto, nos piden que calculemos el número de empleados que tenemos por departamento. Devolved estos resultados en un dataframe. Ordenadas los resultados de mayor a menor basándonos en la cantidad de empleados que tenemos.

```python
df_puestos
```

```python
df_puestos.groupby('id_puesto')['departamento'].count().sort_values(ascending=False).reset_index()
```

6. BONUS Quieren saber que departamentos son los que tienen de media un indice de satisfacción.

💡Pista Para responder a esta pregunta tendremos que unir los dataframes con métodos aprendidos en lecciones anteriores de Pandas.

```python
df_media_bonus = df_puestos.merge(df_historico, on= 'id_empleado')
df_media_bonus.head()
```

```python
df_satisfaccion_dep = df_media_bonus.groupby(['departamento'])['satisfacción'].mean().reset_index()
df_satisfaccion_dep
```


7. BONUS Por último, quieres conocer los 5 empleados que tienen mayor valor en la satisfacción y los 5 empleados que tienen un menor valor de satisfacción. Todo esto respecto a la media. En este caso, conseguiremos una dataframe, donde tengamos el el id del empleado y su satisfacción, pero no sabemos su nombre y appelidos. ¿Se os ocurre como conseguir la información del nombre del empleado?

```python
df_nivel_max_sat = df_media_bonus.groupby(['departamento', 'id_empleado'])['satisfacción'].mean().sort_values(ascending=False).reset_index()
df_nivel_max_sat.head(5)
```

```python
df_nivel_min_sat = df_media_bonus.groupby(['departamento', 'id_empleado'])['satisfacción'].mean().sort_values(ascending=False).reset_index()
df_nivel_min_sat.tail(5)
```

```python
df_nombres_max = df_media_bonus.groupby(['departamento', 'id_empleado', 'nombre', 'apellido'])['satisfacción'].mean().sort_values(ascending=False).reset_index()
df_nombres_max.head(5)
```

```python
df_nombres_min = df_media_bonus.groupby(['departamento', 'id_empleado', 'nombre', 'apellido'])['satisfacción'].mean().sort_values(ascending=False).reset_index()
df_nombres_min.tail(5)
```
